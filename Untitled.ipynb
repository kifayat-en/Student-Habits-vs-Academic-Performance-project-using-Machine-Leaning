# Label Encoding

```python
df.head()
enc = LabelEncoder()
for c in ['gender', 'part_time_job', 'diet_quality', 'parental_education_level', 'internet_quality', 
          'extracurricular_participation']:
    df[c]  = enc.fit_transform(df[c])
```

```python
df.drop('student_id', axis=1, inplace=True)
```

# Data split into Training and Testing sets

```python
X = df.drop('exam_score', axis=1)
y = df['exam_score']

X_train,X_test, y_train,y_test = train_test_split(X,y , test_size=0.2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Models

```python
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=400, n_jobs=100),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=200),
    'Adaboost Regressor': AdaBoostRegressor(),
    "K-Nearest Neighbors": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(),
    'Kmeans Clustering': KMeans()
}
```

```python
summ = []
for name, model in models.items():
    model.fit(X_train,y_train)
    y_preds = model.predict(X_test)    
    score = r2_score(y_test,y_preds)
    mse = mean_squared_error(y_test,y_preds)   
    print(f"\n Name: {name}")
    print(f" R2 Score: {score:.4f}")   
    print(f" MSE: {mse}")
    
    plt.figure(figsize=(9, 5))
    plt.plot(y_test.values, label='Actual', color='black', linewidth=2)
    plt.plot(y_preds, label='Predicted', color='steelblue', linestyle='dashed')
    plt.title(f'{name} - Actual vs Predicted')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    summ.append({
        'Model': name,
        'R2 Score': score,
        'MSE': mse,
    })
    new_df = pd.DataFrame(summ)

# Show all model evaluation metrics in one table
print("\nSummary:")
new_df   
    
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(x='Model', y='MSE', data=new_df)
plt.title('Perfomance of Regression Algorithms (MSE)', fontsize=15)
plt.xticks(rotation=60)
plt.grid()
plt.show()
```
